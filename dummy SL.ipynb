```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, roc_curve
import seaborn as sns
import matplotlib.pyplot as plt

# File paths
train_file_path = "D:/2nd year/Bigdata Analytics/Class room presentation/archive/mitbih_train.csv"  # Replace with your train file path
test_file_path = "D:/2nd year/Bigdata Analytics/Class room presentation/archive/mitbih_test.csv"    # Replace with your test file path

# Load datasets
train_data = pd.read_csv(train_file_path)
test_data = pd.read_csv(test_file_path)

# Rename columns for simplicity
train_data.columns = [f'feature_{i}' for i in range(train_data.shape[1])]
test_data.columns = [f'feature_{i}' for i in range(test_data.shape[1])]

# Fill zero values with previous or next non-zero value
def fill_zeros_with_neighbors(data):
    for col in data.columns:
        data[col] = data[col].replace(0, np.nan)  # Replace zeros with NaN temporarily
        data[col] = data[col].fillna(method='ffill').fillna(method='bfill')  # Fill forward, then backward
    return data

# Apply zero-filling for both train and test datasets
X_train = fill_zeros_with_neighbors(train_data.iloc[:, :-1])
y_train = train_data.iloc[:, -1]

X_test = fill_zeros_with_neighbors(test_data.iloc[:, :-1])
y_test = test_data.iloc[:, -1]

# Train a Random Forest Classifier
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train)

# Predictions
y_pred = clf.predict(X_test)
y_pred_proba = clf.predict_proba(X_test)

# Metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')
auc = roc_auc_score(pd.get_dummies(y_test), y_pred_proba, multi_class='ovr')

# Compute confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
class_labels = np.unique(y_test)

# Class-wise metrics calculation
metrics = []
for i, label in enumerate(class_labels):
    TP = conf_matrix[i, i]
    FP = conf_matrix[:, i].sum() - TP
    FN = conf_matrix[i, :].sum() - TP
    TN = conf_matrix.sum() - (TP + FP + FN)

    TPR = TP / (TP + FN) if (TP + FN) > 0 else 0  # Sensitivity / Recall
    TNR = TN / (TN + FP) if (TN + FP) > 0 else 0  # Specificity
    FNR = FN / (TP + FN) if (TP + FN) > 0 else 0  # Miss Rate
    FPR = FP / (FP + TN) if (FP + TN) > 0 else 0  # Fall-out

    metrics.append({
        'Class': label,
        'TPR (Recall)': TPR,
        'TNR (Specificity)': TNR,
        'FNR': FNR,
        'FPR': FPR,
        'TP': TP,
        'TN': TN,
        'FP': FP,
        'FN': FN
    })

# Convert metrics to a DataFrame for better readability
metrics_df = pd.DataFrame(metrics)

# Display metrics DataFrame
print(metrics_df)

# Bar chart for class-wise metrics
metrics_to_plot = ['TPR (Recall)', 'TNR (Specificity)', 'FNR', 'FPR']
colors = ['green', 'blue', 'red', 'orange']

plt.figure(figsize=(12, 6))
x = np.arange(len(class_labels))
bar_width = 0.2

for i, metric in enumerate(metrics_to_plot):
    plt.bar(x + i * bar_width, metrics_df[metric], width=bar_width, label=metric, color=colors[i])

plt.xticks(x + bar_width, class_labels, fontsize=10)
plt.xlabel('Classes')
plt.ylabel('Scores')
plt.title('Class-wise Metrics')
plt.legend()
plt.show()

# Confusion matrix interpretation for each class
print("\nConfusion Matrix Interpretation for Each Class:")
for i, label in enumerate(class_labels):
    print(f"\nClass {label}:")
    print(f"  TP (True Positives): {metrics_df.loc[i, 'TP']}")
    print(f"  TN (True Negatives): {metrics_df.loc[i, 'TN']}")
    print(f"  FP (False Positives): {metrics_df.loc[i, 'FP']}")
    print(f"  FN (False Negatives): {metrics_df.loc[i, 'FN']}")

# Visualize confusion matrix
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=class_labels, yticklabels=class_labels)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

# Improved Overall Metrics Visualization (Bar Chart)
overall_metrics = {
    "Accuracy": accuracy,
    "Precision": precision,
    "Recall": recall,
    "F1-Score": f1,
    "AUC": auc
}

plt.figure(figsize=(10, 6))
bars = plt.bar(overall_metrics.keys(), overall_metrics.values(), color=['blue', 'green', 'orange', 'red', 'purple'], alpha=0.8)
plt.title('Overall Model Performance Metrics', fontsize=16)
plt.ylabel('Score', fontsize=12)
plt.ylim(0, 1.1)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Annotate each bar with its value
for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.02, f"{bar.get_height():.2f}", ha='center', fontsize=10)

plt.show()
```
