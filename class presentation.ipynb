```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Replace 0 with NaN and fill with forward fill
def preprocess_dataset(file_path):
    """
    Load the dataset, replace 0 with NaN, and apply forward fill.
    """
    print(f"Processing: {file_path}")
    data = pd.read_csv(file_path)
    
    # Replace 0 values with NaN
    data = data.replace(0, float('nan'))
    
    # Forward fill NaN values
    data = data.ffill(axis=0)
    
    print(f"Processed dataset: {file_path}\n")
    return data

# Visualize label distribution
def plot_label_distribution(data, title):
    """
    Plot the distribution of labels in the dataset.
    """
    print("Available columns:", data.columns)
    
    # Check if "Label" or last column exists and is valid
    label_col = data.columns[-1]  # Assuming the last column is the label column
    if label_col not in data.columns or data[label_col].isnull().all():
        print(f"Error: Could not find a valid label column in the dataset.")
        return
    
    plt.figure(figsize=(8, 5))
    sns.countplot(x=label_col, data=data, palette='Set2')
    plt.title(title)
    plt.xlabel('Label')
    plt.ylabel('Count')
    plt.show()

# Visualize a single signal in ECG-like format
def plot_ecg_signal(data, row_idx, title):
    """
    Plot a single signal in an ECG-like format for a given row index.
    """
    plt.figure(figsize=(12, 6))
    sample_signal = data.iloc[row_idx, :-1]
    time_points = range(len(sample_signal))  # Create a time axis
    
    plt.plot(time_points, sample_signal, color='blue', linewidth=1)
    plt.title(title)
    plt.xlabel('Time (ms)')
    plt.ylabel('Amplitude (mV)')
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.axhline(y=0, color='black', linewidth=0.8)  # Baseline
    plt.tight_layout()
    plt.show()

# File paths for datasets
datasets = {
    "MIT-BIH Train": "D:/2nd year/Bigdata Analytics/Class room presentation/archive/mitbih_train.csv",
    "MIT-BIH Test": "D:/2nd year/Bigdata Analytics/Class room presentation/archive/mitbih_test.csv",
    "PTB Abnormal": "D:/2nd year/Bigdata Analytics/Class room presentation/archive/ptbdb_abnormal.csv",
    "PTB Normal": "D:/2nd year/Bigdata Analytics/Class room presentation/archive/ptbdb_normal.csv"
}

# Process and visualize each dataset
for dataset_name, file_path in datasets.items():
    # Check if file exists
    if os.path.exists(file_path):
        # Load and preprocess the dataset
        data = preprocess_dataset(file_path)
        
        # Save the cleaned dataset
        cleaned_file_path = file_path.replace(".csv", "_cleaned.csv")
        data.to_csv(cleaned_file_path, index=False)
        print(f"Cleaned dataset saved: {cleaned_file_path}\n")
        
        # 1. Visualize label distribution
        plot_label_distribution(data, f"Label Distribution in {dataset_name} Dataset")
        
        # 2. Visualize a single signal in ECG-like format (first row)
        plot_ecg_signal(data, row_idx=0, title=f"ECG-Like Signal for Row 0 in {dataset_name} Dataset")
    else:
        print(f"File not found: {file_path}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score, 
                             roc_auc_score, confusion_matrix, roc_curve, auc)

# Load datasets
train_file_path = "D:/2nd year/Bigdata Analytics/Class room presentation/archive/mitbih_train.csv"  # Update with your actual path
test_file_path = "D:/2nd year/Bigdata Analytics/Class room presentation/archive/mitbih_test.csv"    # Update with your actual path

train_data = pd.read_csv(train_file_path)
test_data = pd.read_csv(test_file_path)

# Rename columns for simplicity
train_data.columns = [f'feature_{i}' for i in range(train_data.shape[1])]
test_data.columns = [f'feature_{i}' for i in range(test_data.shape[1])]

# Prepare features (X) and target (y) for training and testing
X_train = train_data.iloc[:, :-1]
y_train = train_data.iloc[:, -1]

X_test = test_data.iloc[:, :-1]
y_test = test_data.iloc[:, -1]

# Train a Random Forest Classifier
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train)

# Predictions
y_pred = clf.predict(X_test)
y_pred_proba = clf.predict_proba(X_test)

# Metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')
auc_score = roc_auc_score(pd.get_dummies(y_test), y_pred_proba, multi_class='ovr')

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Visualization setup
fig = plt.figure(figsize=(18, 15))

# 1. Confusion Matrix Visualization
ax1 = fig.add_subplot(221)
sns.heatmap(conf_matrix, annot=False, fmt='d', cmap='coolwarm', 
            xticklabels=['Normal', 'Abnormal 1', 'Abnormal 2', 'Abnormal 3', 'Abnormal 4'],
            yticklabels=['Normal', 'Abnormal 1', 'Abnormal 2', 'Abnormal 3', 'Abnormal 4'], 
            cbar_kws={'label': 'Frequency'}, ax=ax1)

row_sums = conf_matrix.sum(axis=1, keepdims=True)
conf_matrix_percent = conf_matrix / row_sums * 100  # Convert to percentages
for i in range(conf_matrix.shape[0]):
    for j in range(conf_matrix.shape[1]):
        ax1.text(j + 0.5, i + 0.4, f"{conf_matrix_percent[i, j]:.1f}%", 
                 ha='center', va='center', color='black', fontsize=10)
        ax1.text(j + 0.5, i + 0.6, f"{conf_matrix[i, j]}", 
                 ha='center', va='center', color='white', fontsize=12, fontweight='bold')
ax1.set_title('Confusion Matrix with Frequencies and Percentages', fontsize=16)
ax1.set_xlabel('Predicted Labels')
ax1.set_ylabel('True Labels')

# 2. Bar Graph for Metrics
metrics = {
    "Accuracy": accuracy,
    "Precision": precision,
    "Recall": recall,
    "F1-Score": f1,
    "AUC": auc_score
}
colors = ['#5b9bd5', '#ed7d31', '#a5a5a5', '#ffc000', '#70ad47']
ax2 = fig.add_subplot(222)
ax2.bar(metrics.keys(), metrics.values(), color=colors, alpha=0.85, edgecolor='black')
ax2.set_title('Model Performance Metrics', fontsize=16)
ax2.set_ylabel('Score')
ax2.set_ylim(0, 1.1)
for index, value in enumerate(metrics.values()):
    ax2.text(index, value + 0.02, f"{value:.2f}", ha='center', fontsize=10)

# 3. Radar Chart for Metrics
categories = list(metrics.keys())
values = list(metrics.values())
values += values[:1]  # Repeat the first value to close the radar chart

angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
angles += angles[:1]  # Ensure the same closing angle for the radar chart

ax3 = fig.add_subplot(223, polar=True)  # Declare this subplot as polar
ax3.plot(angles, values, marker='o', linestyle='-', linewidth=2, color='b', alpha=0.75)
ax3.fill(angles, values, color='blue', alpha=0.25)
ax3.set_yticks(np.linspace(0, 1, 6))  # Set radial ticks
ax3.set_xticks(angles)
ax3.set_xticklabels(categories, fontsize=10)  # Include all categories for labels
ax3.set_title('Radar Chart of Metrics', fontsize=16)

# 4. ROC Curves
ax4 = fig.add_subplot(224)
n_classes = len(pd.get_dummies(y_test).columns)
for i in range(n_classes):
    fpr, tpr, _ = roc_curve(pd.get_dummies(y_test).iloc[:, i], y_pred_proba[:, i])
    roc_auc = auc(fpr, tpr)
    ax4.plot(fpr, tpr, label=f'Class {i} (AUC: {roc_auc:.2f})')

ax4.plot([0, 1], [0, 1], 'k--', lw=1)
ax4.set_title('ROC Curves for Each Class', fontsize=16)
ax4.set_xlabel('False Positive Rate')
ax4.set_ylabel('True Positive Rate')
ax4.legend(loc='lower right')

# Final adjustments
plt.tight_layout()
plt.show()

# Print final metrics
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-Score: {f1:.2f}")
print(f"AUC: {auc_score:.2f}")
```
